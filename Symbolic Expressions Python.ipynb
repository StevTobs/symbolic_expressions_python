$ f(x) = \lceil  x \rceil$

$ g(x) = \lfloor  x \rfloor$

$ h(x) = x - \lfloor  x \rfloor$

````python
f = ceiling(x) #return the smallest interger valie which is not less than x
g = floor(x) #return the largest interger value which is not greater than x
h = frac(x) #return the fractional part of x
````

````python
fx = lambdify(x,f,modules=['numpy'])
gx = lambdify(x,g,modules=['numpy'])
hx = lambdify(x,h,modules=['numpy'])
````

````python
xVals = linspace(-10,10,100)
plt.plot(xVals, fx(xVals),label = ['f(x)=ceiling(x)'])
plt.plot(xVals, gx(xVals),label = ['g(x)=floor(x)'])
plt.plot(xVals, hx(xVals),label = ['h(x)=frac(x)'])
plt.xlabel('x')
plt.ylabel('f(x),g(x),h(x)')
plt.title('Inverse Hyperbolic sines and cosines')
plt.legend()
plt.show()
````

$$\frac{x^{3} + x^{2} - x - 1}{x^{2} + 2 x + 1}$$

````python
f = (x**3 + x**2-x-1)/( x**2 + 2*x +1 )
````

$$f(x) = \frac{x^{3} + x^{2} - x - 1}{x^{2} + 2 x + 1} = \frac{(x-1)(x+1)^2}{(x+1)^2} = x-1$$





````python
f.simplify()
````

$$\frac{\Gamma\left(x\right)}{\Gamma\left(x - 5\right)}$$
$$\Gamma (x) = x\Gamma (x-1) =x(x-1)\Gamma (x-2) =...=x(x-1)(x-2)(x-3)(x-4)\Gamma (x-5)$$

$$f(x) = {\Gamma (x) \over \Gamma( x-5)} = {x(x-1)(x-2)(x-3)(x-4)\Gamma (x-5) \over \Gamma( x-5) } = x(x-1)(x-2)(x-3)(x-4)$$

````python
f = gamma(x) / gamma(x-5)
````

````python
pprint(f)
````

````python
f.simplify()
````

````python
simplify(f)
````
